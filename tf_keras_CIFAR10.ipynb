```python
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.utils import to_categorical        ######## import necessary libraries from tensorflow.keras
```

```python
from tensorflow.keras.datasets import cifar10            ###### import CIFAR-10 dataset
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

```python
# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
train_images = train_images.reshape((50000, 32, 32, 3))    ##### train dataset size is 50000, shape is 32*32 and 3 for color image
train_images = train_images.astype('float32') / 255        ###### we convert train image as float values
test_images = test_images.reshape((10000, 32, 32, 3))      ##### test dataset size is 10000, shape is 32*32 and 3 for color image
test_images = test_images.astype('float32') / 255           ###### we convert test image as float values
train_labels = to_categorical(train_labels)                 ##### we convert output as categorical
test_labels = to_categorical(test_labels)                    ##### we convert output as categorical
```

```python
from tensorflow.keras import layers            ##### import layers
from tensorflow.keras import models            ##### import models
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))  #### we add Convelutional2D layer and we use 32 as filter size and 3 for color images
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))       ##### we use Relu Activation function for output layer
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.summary()

model.add(layers.Flatten())                                       # it converts 2D to 1D
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])              ##### we compile model as optimizer adam and loss= categorical_crossentropy
model.fit(train_images, train_labels, epochs=5, batch_size=64, verbose=2, validation_split=0.2)     ##### we train a model on train dataset
test_loss, test_acc = model.evaluate(test_images, test_labels)                                      #### Evaluate the test accuracy
test_acc                                                                                            #### test accuracy is 100%
```

```python
#Evaluating on the train data
train_acc = model.evaluate(train_images, train_labels, verbose=2)             #### Evaluate the train accuracy
train_acc                                                                     #### train accuracy is 100%
```
